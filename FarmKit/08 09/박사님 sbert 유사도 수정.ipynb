```python
# Sentence Bert 
from sentence_transformers import SentenceTransformer, util
import torch
# model = SentenceTransformer('all-MiniLM-L6-v2')
model = SentenceTransformer('all-mpnet-base-v2')
```

```python
from sklearn.metrics.pairwise import cosine_similarity, euclidean_distances

# 재료만 embedding
print('food_ingres embedding started ... ')
ingres_embeddings = model.encode(food_ingres)

# Description 만 embedding
print('food_descs embedding started ... ')
descs_embeddings = model.encode(food_descs)

# 설명 + 재료의 embedding
print('food_ingres_descs embedding started ... ')
ingres_descs_embeddings = model.encode(food_ingres_descs)

# 유사도 계산해보기 ... 
ingres_scores_dist = euclidean_distances(ingres_embeddings, ingres_embeddings)
descs_scores_dist = euclidean_distances(descs_embeddings, descs_embeddings)
ingres_descs_scores_dist = euclidean_distances(ingres_descs_embeddings, ingres_descs_embeddings)
# ingres_scores_dist = torch.cdist(torch.tensor(ingres_embeddings), torch.tensor(ingres_embeddings))
# descs_scores_dist = torch.cdist(torch.tensor(descs_embeddings), torch.tensor(descs_embeddings))
# ingres_descs_scores_dist = torch.cdist(torch.tensor(ingres_descs_embeddings), torch.tensor(ingres_descs_embeddings))
ingres_scores = cosine_similarity(ingres_embeddings, ingres_embeddings)
descs_scores = cosine_similarity(descs_embeddings, descs_embeddings)
ingres_descs_scores = cosine_similarity(ingres_descs_embeddings, ingres_descs_embeddings)
# ingres_scores = util.cos_sim(ingres_embeddings, ingres_embeddings)
# descs_scores = util.cos_sim(descs_embeddings, descs_embeddings)
# ingres_descs_scores = util.cos_sim(ingres_descs_embeddings, ingres_descs_embeddings)
```

```python
import pandas as pd
import random
```

```python
def get_top(idx, scores, sortReverse) :
    pair = []
    for i in range(len(scores)):
        pair.append({'index':i, 'score':scores[idx][i]})
    pair = sorted(pair, key=lambda x: x['score'], reverse=sortReverse)
    ret = []
    for i in range(20):
        ret.append(food_names[pair[i]['index']].replace('\n', ''))
    return ret

def to_dataframe (ing_cos, ing_euc, des_cos, des_euc, id_cos, id_euc):
    df = pd.DataFrame()
    df['ingredients (cos)'] = ing_cos
    df['ingredients (euc)'] = ing_euc
    df['description (cos)'] = des_cos
    df['description (euc)'] = des_euc
    df['ing+des (cos)'] = id_cos
    df['ing+des (euc)'] = id_euc
    return df

def food_recommendation(idx) : 
    print(food_names[idx], idx)
    df = to_dataframe(get_top(idx, ingres_scores, True), get_top(idx, ingres_scores_dist, False), get_top(idx, descs_scores, True), get_top(idx, descs_scores_dist, False), get_top(idx, ingres_descs_scores, True), get_top(idx, ingres_descs_scores_dist, False))
    return df
```
